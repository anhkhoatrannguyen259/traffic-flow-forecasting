```python
import numpy as np
import matplotlib.pyplot as plt

labels = ["1/1-14/3","15/3-26/5",	"27/5-7/8",	"8/8-19/10"]
rmse_lstm = [213.0477904,	155.569203,	151.2280377,	147.2753369]
smape_lstm = [0.3218768076,	0.2929862306,	0.2831104816,	0.2814376416]
rmse_arima = [1342.315697,	1438.030331,	1418.771447,	1317.655028]
smape_arima = [1.662540902,	1.850994591,	1.817970947,	1.419868197]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, (ax, ax1) = plt.subplots(figsize=(14.5,8.5), nrows=1, ncols=2, dpi=300)
rects1 = ax.bar(x - width/2, rmse_lstm, width, label='LSTM')
rects2 = ax.bar(x + width/2, rmse_arima, width, label='ARIMA')

# Set general font size
plt.rcParams['font.size'] = '15'

# Set tick font size for ax
for label in (ax.get_xticklabels() + ax.get_yticklabels()):
	label.set_fontsize(15)
 
 # Set tick font size for ax1
for label2 in (ax1.get_xticklabels() + ax1.get_yticklabels()):
	label2.set_fontsize(15)

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('RMSE')
ax.set_xlabel("Range of training data")
ax.set_title('Compare RMSE value of models')
ax.set_xticks(np.arange(len(labels)))
ax.set_xticklabels(labels, rotation=35)
#plt.xticks(np.arange(11), labels, rotation=45)
#plt.grid(True)
ax.legend(loc=0, prop={'size': 11})

def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.009*height,
                '%d' % float(height),
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

#-----------------------------------------------------------------------------------
#-----------------------------------------------------------------------------------

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

#fig, ax = plt.subplots(figsize=(10,6))
rects1 = ax1.bar(x - width/2, smape_lstm, width, label='LSTM')
rects2 = ax1.bar(x + width/2, smape_arima, width, label='ARIMA')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax1.set_ylabel('sMAPE')
ax1.set_xlabel("Range of training data")
ax1.set_title('Compare sMAPE value of models')
ax1.set_xticks(np.arange(len(labels)))
ax1.set_xticklabels(labels, rotation=35)
#plt.xticks(np.arange(11), labels, rotation=45)
#plt.grid(True)
ax1.legend(loc=0, prop={'size': 11})

def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax1.text(rect.get_x() + rect.get_width()/2., 1.009*height,
                '%.2f' % float(height),
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()
plt.savefig("compare_by_timeliness_ver2.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
