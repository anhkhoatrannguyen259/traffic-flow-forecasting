```python
data_historical = pd.read_csv('/content/file.csv')
data_historical['Visibility'] = data_historical['Visibility']*1000
data_historical['Visibility'] = data_historical['Visibility'].astype('int64')

data_historical['Date time'] = data_historical['Date time'].astype('datetime64')
data_historical['Sunrise'] = data_historical['Sunrise'].astype('datetime64')
data_historical['Sunset'] = data_historical['Sunset'].astype('datetime64')

data_historical['Relative Humidity'] = round(data_historical['Relative Humidity'],0)
data_historical['Relative Humidity'] = data_historical['Relative Humidity'].astype('int64')
data_historical['Relative Humidity'].astype('int64')
data_historical['sun'] = np.nan
data_historical
```

```python
for i in range(len(data_historical)):
  if data_historical['Date time'][i].hour == data_historical['Sunrise'][i].hour:
    data_historical['sun'][i] = round((60 - data_historical['Sunrise'][i].minute)/60, 1)
  elif data_historical['Date time'][i].hour == data_historical['Sunset'][i].hour:
    data_historical['sun'][i] = round(data_historical['Sunset'][i].minute/60, 1)
  elif data_historical['Date time'][i] > data_historical['Sunrise'][i] and data_historical['Date time'][i] < data_historical['Sunset'][i]:
    data_historical['sun'][i] = 1
  else:
    data_historical['sun'][i] = 0
```

## Merge historical weather data with traffic data for demo

```python
traffic_data_demo.info()
```

```python
traffic_data_demo['datetime'] = traffic_data_demo['datetime'].astype('datetime64')
traffic_data_demo.info()
```
