```python
location_df = pd.read_csv("/content/drive/MyDrive/Khoá luận tốt nghiệp/Dataset/Location_Official.csv")
# location_df = location_df[:2]
location_df
```

```python
df = pd.DataFrame()

dict_date = {"thang1": ["2021-01-01", "2021-01-31"],
             "thang2": ["2021-02-01", "2021-02-28"],
             "thang3": ["2021-03-01", "2021-03-31"],
             "thang4": ["2021-04-01", "2021-04-30"],
             "thang5": ["2021-05-01", "2021-05-31"],
             "thang6": ["2021-06-01", "2021-06-30"],
             "thang7": ["2021-07-01", "2021-07-31"],
             "thang8": ["2021-08-01", "2021-08-31"],
             "thang9": ["2021-09-01", "2021-09-30"],
             "thang10": ["2021-10-01", "2021-10-31"],
             "thang11": ["2021-11-01", "2021-11-30"],
             "thang12": ["2021-12-01", "2021-12-31"],
             }

for i in range(len(location_df)):
  location = location_df["Location"][i]
  sgid_spid = location_df["SgidSpid"][i]
  print("Location " + str(i) + ": " + location_df["Site Name"][i])

  for k in dict_date:
    print(" ----- " + k)
    file_url = "https://trafficdata.tii.ie/tfdaysreport.asp?{}&reportdate={}&enddate={}&dir=0&excel=1".format(sgid_spid, dict_date[k][0], dict_date[k][1])
    file_data = requests.get(file_url).content
    # create the file in write binary mode, because the data we get from net is in binary
    with open("file.xls", "wb") as file:
        file.write(file_data)
    data = pd.read_html('file.xls')
    df2 = data[5]
    id = data[0][1][1]
    df2 = df2.iloc[:24, :-3]
    df2.columns = [x[1] for x in df2.columns]
    df2 = df2.melt(id_vars=['Time'], var_name=['Date'])
    df2["datetime"] = df2["Date"] + " " + df2["Time"]
    df2 = df2[["datetime", "value"]]
    df2["datetime"] = df2["datetime"].astype("datetime64")
    # df2["value"] = df2["value"].astype("int64")
    df2["id"] = id
    df2["location"] = location
    df2 = df2[["id", "datetime", "location", "value"]]
    df = pd.concat([df, df2], ignore_index=True)

df    
```
