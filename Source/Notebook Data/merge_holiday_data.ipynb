```python
data = pd.read_csv("/content/drive/MyDrive/Courses/Khoá luận tốt nghiệp/Dataset/final_data_5.csv")
data['datetime'] = data['datetime'].astype('datetime64')
data
```

```python
data.info()
```

```python
data['holiday'] = data['datetime'].dt.date
data['holiday'] = data['holiday'].astype('str')
data
```

```python
data.info()
```

```python
list_holiday = ['2021-01-01', '2021-03-17', '2021-04-05', '2021-05-03', '2021-06-07', '2021-08-02', '2021-10-25', '2021-12-25', '2021-12-26']

for i in range(len(data)):
  if data['holiday'][i] in list_holiday:
    data['holiday'][i] = 1
  else:
    data['holiday'][i] = 0
```

```python
data.drop('location',axis=1, inplace=True)
data
```
